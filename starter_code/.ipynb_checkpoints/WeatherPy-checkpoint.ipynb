```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time
from scipy.stats import linregress
import openweathermapy as ow


# Import API key
from api_keys import weather_api_key

# Incorporated citipy to determine city based on latitude and longitude
from citipy import citipy

# Output File (CSV)
output_data_file = "output_data/cities.csv"

# Range of latitudes and longitudes
lat_range = (-90, 90)
lng_range = (-180, 180)
```

## Generate Cities List

```python
# List for holding lat_lngs and cities
lat_lngs = []
cities = []

# Create a set of random lat and lng combinations
lats = np.random.uniform(low=-90.000, high=90.000, size=1500)
lngs = np.random.uniform(low=-180.000, high=180.000, size=1500)
lat_lngs = zip(lats, lngs)

# Identify nearest city for each lat, lng combination
for lat_lng in lat_lngs:
    city = citipy.nearest_city(lat_lng[0], lat_lng[1]).city_name
    
    # If the city is unique, then add it to a our cities list
    if city not in cities:
        cities.append(city)

# Print the city count to confirm sufficient count
len(cities)
```

### Perform API Calls
* Perform a weather check on each city using a series of successive API calls.
* Include a print log of each city as it'sbeing processed (with the city number and city name).


```python
#base URL and keys to call from data frame
base_url = "http://api.openweathermap.org/data/2.5/weather?units=Imperial&APPID=" + weather_api_key
setting = {'units':'Imperial','appid':weather_api_key}
desired_keys = ['clouds.all','sys.country','dt','main.humidity','main.temp_max','wind.speed','coord.lat','coord.lon']


# Create new lists for data frame
cloudiness = []
country = []
date = []
humidity = []
max_temp = []
wind_speed = []
lat =[]
lng =[]
cities_list = []

#Print beginning and start count
print('Beginning Data Retrieval')
print('---------------------')
count = 1 
err=0
#loop through ciities in list and grab data to append lists
for city in cities:
    city_url = base_url + '&q=' + city.replace(' ','+')
    print(f'Processing record # {count} | {city}')
   # print(city_url)
    try:
        city_data = ow.get_current(city,**setting)
        cl,ct,dt,hm,mt,ws,lt,ln = city_data(*desired_keys)
        cloudiness.append(cl)
        country.append(ct)
        date.append(dt)
        humidity.append(hm)
        max_temp.append(mt)
        wind_speed.append(ws)
        lat.append(lt)
        lng.append(ln)
        cities_list.append(city)
    except Exception as e:
        print('City not found. skipping...')
        err=err+1
    count = count+1
    
    
#print errors and end

print("end of loop")
print(f'Could not get data for {err} cities')
```

### Convert Raw Data to DataFrame
* Export the city data into a .csv.
* Display the DataFrame

```python
# create data frane count records in data frame
weather_data = pd.DataFrame({
   'City':cities_list,'Cloudiness':cloudiness,'Country':country, 'Date':date,'Humidity':humidity,'Latitude':lat,
    'Longitude':lng,'Max Temperature':max_temp,'Wind Speed':wind_speed,})
weather_data.count()
```

```python
#save data frame in a csv file
weather_data.to_csv("weather_data.csv")
```

## Linear Regression

```python
# OPTIONAL: Create a function to create Linear Regression plots
```

```python
# Create Northern Hemisphere DataFrames
Northern = weather_data[weather_data["Latitude"] >= 0]
Northern
```

```python
#Create Southern DataFrame
Southern = weather_data[weather_data["Latitude"] < 0]
Southern
```

####  Northern Hemisphere - Max Temp vs. Latitude Linear Regression

```python
#import scripy set columns for variables lat north and max temp north
import scipy.stats as st
MaxTempNorth = Northern.iloc[:,7]
LatNorth = Northern.iloc[:,5]

#Run statistical analysis on two variables
(slope, intercept, rvalue, pvalue, stderr) = st.linregress(MaxTempNorth, LatNorth)

#Create regress line formula print correlation and plot with regress line and equation
regress_values = MaxTempNorth * slope + intercept
print(f"The r-squared is {rvalue}")
plt.scatter(MaxTempNorth,LatNorth)
plt.plot(MaxTempNorth,regress_values,"r-")
line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))
plt.annotate(line_eq,(2,1),fontsize=12,color="red")
plt.title("Northern Hemisphere Max Temp vs Latitude")
plt.xlabel('Max Temp (F)')
plt.ylabel('Latitude')
plt.show()
```

```python
MaxTempSouth = Southern.iloc[:,7]
LatSouth = Southern.iloc[:,5]

(slope, intercept, rvalue, pvalue, stderr) = st.linregress(MaxTempSouth, LatSouth)
print(f"The r-squared is {rvalue}")
regress_values = MaxTempSouth * slope + intercept
plt.scatter(MaxTempSouth,LatSouth)
plt.plot(MaxTempSouth,regress_values,"r-")
line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))
plt.annotate(line_eq,(80,-50),fontsize=12,color="red")
plt.title("Southern Hemisphere Max Temp vs Latitude")
plt.xlabel('Max Temp (F)')
plt.ylabel('Latitude')
#plt.show()
```
